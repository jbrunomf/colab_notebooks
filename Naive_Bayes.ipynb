```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, classification_report

iris = load_iris()

X_train, X_test, y_train, y_test = train_test_split(
    iris.data, iris.target, random_state=42
)
```

# Gaussian NB é capaz dei nduzir um modelo a partir de dados numéricos.
Os três passos para indiçãod e um modelo são:
- Instanciar o modelo GaussianNB()
- Treinar o modelo: fit()
- Testar o modelo: predict()

```python
gnb = GaussianNB()
y_pred = gnb.fit(X_train, y_train).predict(X_test)

print("Número de erros de classificação: {0} de {1}".format((y_test != y_pred).sum(), iris.data.shape[0]))

ac = gnb.score(X_test, y_test)

print("\nAcurácia do modelo: {0: .2f}%\n".format(100*ac))

print(classification_report(y_test, y_pred, target_names=iris.target_names))

cnf_matrix = confusion_matrix(y_test, y_pred)
cnf_table = pd.DataFrame(data=cnf_matrix, index=iris.target_names, columns=[x + "(prev)" for x in iris.target_names])
print(cnf_table)
```
